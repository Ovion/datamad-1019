```python
%matplotlib inline
import matplotlib.pyplot as plt
import math

Fm = 4
f = 1 ## Hz
T = 1/f
print(f'Freq:{f} Periodo:{T}')
print(f'Freq muestreo:{Fm} Periodo muestreo:{1/Fm}')

x = np.arange(0, 1+(1/Fm), 1/Fm)
y = np.sin(2*math.pi*f*x)

print(f'muestras:{len(x)}')

plt.plot(x,y, 'o-')
plt.ylabel('Amplitid')
plt.xlabel('tiempo')
```

```python

def datos(Fm,fosc, Amplitude=1):
    x = np.arange(0, 1+(1/Fm), 1/Fm)
    y = Amplitude * np.sin(2*math.pi*fosc*x)
    return x,y

Fm = 20
x1,y1 = datos(Fm,2)
x2,y2 = datos(Fm*2,4,0.5)
plt.plot(x1,y1,'-o')
plt.plot(x2,y2,'-o')

print(len(y2),len(y1))
alarma = (y2-y1)>0 # Cuando y2 sea mas grande que y1, disparamos la alarma
plt.plot(x2,alarma)


plt.ylabel('Amplitid')
plt.xlabel('tiempo')
plt.legend(["SensorA","SensorB","Disparador alarma"])
```

```python
from scipy.interpolate import interp1d

x = np.arange(0.0,11.0,1)
y = np.sin(2*math.pi*(1/10)*x)
plt.plot(x,y,'-o')


f = interp1d(x, y)
xnew = np.arange(0.0, 10.0, 0.5)
print(len(xnew),len(x))
print(x)
print(xnew)
plt.plot(xnew,1+f(xnew),'-o')
```

```python

def datos(Fm,fosc, Amplitude=1):
    x = np.arange(0, 1+(1/Fm), 1/Fm)
    y = Amplitude * np.sin(2*math.pi*fosc*x)
    return x,y

Fm = 20
x1,y1 = datos(Fm,2)
x2,y2 = datos(Fm*2,4,0.5)
plt.plot(x1,y1,'-o')
plt.plot(x2,y2,'-o')

print(x1)
print(x2)
print(len(y2),len(y1))

f = interp1d(x1,y1)

z = f(x2)
print('lens',len(z),len(y2))
alarma = (y2-z)>0 # Cuando y2 sea mas grande que y1, disparamos la alarma
plt.plot(x2,alarma)


plt.ylabel('Amplitid')
plt.xlabel('tiempo')
plt.legend(["SensorA","SensorB","Disparador alarma"])
```
