```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html
import pandas as pd

df = pd.read_csv("./data/vehicles/vehicles.csv")
df.head()
```

```python
data = df[["Make","Model","Year","Cylinders","Engine Displacement"]]
data.head()
```

```python
# Apply sobre una serie
def transformFn(e):
    return e*3
  
serie = data.head(10)["Make"]
transformedSerie = serie.apply(transformFn)
display(transformedSerie)
data["TR"] = transformedSerie
display(data)
```

```python
# Apply sobre un dataframe
def transformFn(e):
    print(e)
    return "{} {}".format(e["Make"],e["Model"])
  
df2 = data.head(10)[["Make","Model"]]
tr = df2.apply(transformFn, axis=0)
display(tr)
```

```python
# Apply sobre un dataframe con broadcast
def transformFn(e):
    return "{} {}".format(e["Make"],e["Model"])
  
df2 = data.head(10)[["Make","Model"]]
tr = df2.apply(transformFn, axis=1, result_type="broadcast")
display(tr)
```

```python
# Apply sobre un dataframe con broadcast
def transformFn(e):
    print(e)
#    return "{} {}".format(e["Make"],e["Model"])
    return (0,1,2)
  
df2 = data.head(10)[["Make","Model"]]
tr = df2.apply(transformFn, axis=1, result_type='expand')
display(tr)
```

```python
def splitTr(e):
    return e["Transmission"].split(' ')

x = df[["Make","Transmission"]]
result = x.apply(splitTr, axis=1, result_type="expand")
result.columns = ["Trans", "Marchas"]
```

```python
pd.concat([x,result], axis=1)
```
