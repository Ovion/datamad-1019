```python
import matplotlib.pyplot as plt
from sklearn import datasets, svm, metrics

# The digits dataset
digits = datasets.load_digits()
```

```python
digits["target"]
```

```python
print(len(digits["data"]))
number = digits["data"][0].reshape((8,8))
```

```python
for i,number in enumerate(digits["data"][0:20]):
    number = number.reshape((8,8))
    plt.subplot(5,4,i+1)
    plt.imshow(number, cmap='gray')
    plt.title(digits["target"][i])
```

```python
import pandas as pd
X = pd.DataFrame(digits["data"])
#df["GT"] = digits["target"]
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
XReduced = pd.DataFrame(pca.fit_transform(X))
```

```python
XReduced["GT"] = digits["target"]
```

```python
only_100 = XReduced[:20]
plt.plot(only_100[0], only_100[1],'o')

for idx, point in only_100.iterrows():
#    print(point[1])
#    print(point)
    plt.annotate(str(point["GT"]), (point[0], point[1]), fontsize=10)
```

```python
only_100 = XReduced[:1000]
plt.scatter(only_100[0], only_100[1], c=only_100["GT"], alpha=0.8, edgecolors='none')
```
